```python
def printTime(strt, fin):
    '''print execution times '''
    totalTime = fin - strt
    totalTime= round(totalTime, 0)
    if totalTime < 60.0:
        print(f'completed: {totalTime} second(s)')
    elif totalTime >= 60.0:
        minutes = totalTime//60
        seconds = totalTime - (minutes * 60)
        seconds = round(seconds, 2)
        print(f'completed: {minutes} minutes {seconds} second(s)')

def getTensorflow():
    %pip install tensorflow
    print()
    sleep(1)
    return f'installed: {C.BIBlue}tensorflow{C.ColorOff}'

def getTFHub():
    %pip install --upgrade tensorflow-hub
    print()
    sleep(1)
    return f'installed: {C.BIBlue}tensorflow-hub{C.ColorOff}'
    
def getTFLite():
    %pip install tflite-model-maker
    print()
    sleep(1)
    return f'installed: {C.BIBlue}tflite-model-maker{C.ColorOff}'

def getNumpy():
    %pip install numpy
    print()
    sleep(1)
    return f'installed: {C.BIBlue}numpy{C.ColorOff}'
    
def getCV2():
    %pip install opencv-python-headless # works.
    print()
    sleep(1)
    return f'installed: {C.BIBlue}opencv-python-headless{C.ColorOff}'
    # %pip install opencv-contrib-python-headless

def getMatplotlib():
    %pip install matplotlib
    print()
    sleep(1)
    return f'installed: {C.BIBlue}matplotlib{C.ColorOff}'
    
def getSciPi():
    %pip install SciPy
    print()
    sleep(1)
    return f'installed: {C.BIBlue}SciPy{C.ColorOff}'

def getResults():
    %pip install --upgrade results
    print()
    sleep(1)
    return f'installed: {C.BIBlue}results{C.ColorOff}'

def getBeautifulSoup():
    ''' '''
    %pip install beautifulsoup4
    sleep(1)
    return f'installed: {C.BIBlue}beautifulsoup4{C.ColorOff}'

def get_htmlib():
    '''USED WITH BeautifulSoup'''
    %pip install html5lib
    sleep(1)
    return f'installed: {C.BIBlue}html5lib{C.ColorOff}'

def get_lxml():
    '''USED WITH BeautifulSoup'''
    %pip install lxml
    sleep(1)
    return f'installed: {C.BIBlue}lxml{C.ColorOff}'
print('ok')
```

```python
from concurrent.futures import ThreadPoolExecutor

start = perf_counter()
with ThreadPoolExecutor(max_workers=2) as executor:
    thread_1 = executor.submit(getTensorflow)
    thread_2 = executor.submit(getTFHub)
    thread_3 = executor.submit(getTFLite)
    thread_4 = executor.submit(getNumpy)
    thread_5 = executor.submit(getMatplotlib)
    thread_6 = executor.submit(getCV2)
    thread_7 = executor.submit(getSciPi)
    thread_8 = executor.submit(getResults)
    thread_9 = executor.submit(getBeautifulSoup)
    thread_10 = executor.submit(get_htmlib)
    thread_11 = executor.submit(get_lxml)
finish = perf_counter()
printTime(start,finish)
```
